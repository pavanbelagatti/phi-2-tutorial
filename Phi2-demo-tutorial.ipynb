## Importing Necessary Libraries

Import the required Python libraries for your task. Place these import statements at the top of your notebook

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
```

## Load Tokenizer and Model

load the Phi-2 model and its corresponding tokenizer. This step involves downloading the model from Hugging Face, which can be large and may take some time depending on your internet connection.

```python
tokenizer = AutoTokenizer.from_pretrained(
    "microsoft/phi-2",
    trust_remote_code=True  # Ensure you trust the source!
)

model = AutoModelForCausalLM.from_pretrained(
    "microsoft/phi-2",
    torch_dtype=torch.float32,  # Use full precision
    device_map="auto",
    trust_remote_code=True  # Ensure you trust the source!
)
```

## Define Your Prompt

#### Specify the prompt you want to use for generation. This is what you will feed into the model to generate text.

```python
prompt = "Give me a list of 7 wonders of the world."
```

## Generating Text

#### This cell will handle the actual text generation:

```python
with torch.no_grad():
    token_ids = tokenizer.encode(prompt, add_special_tokens=True, return_tensors="pt")
    output_ids = model.generate(
        token_ids.to(model.device),
        max_new_tokens=100,  # Adjust as needed
        do_sample=True,
        temperature=0.3  # Adjust for creativity
    )

output = tokenizer.decode(output_ids[0], skip_special_tokens=True)
print(output)
```
